```python
import matplotlib.pyplot as plt

plt.figure(figsize=[10, 200])  # Установите размер фигуры в начале
i = 0
for el in df2.columns:
    if el not in ['stage_4_output_danger_gas', 'DateTime']:
        i += 1
        plt.subplot(22, 1, i)  # Установите подграфик
        plt.scatter(df2[el], df2['stage_4_output_danger_gas'])
        plt.xlabel(el)
        plt.ylabel('stage_4_output_danger_gas')

```
